```python
#New column for CLose column
data=df.filter(['Close'])
#Converting Series data to numpy array
dataset=data.values
```

```python
df.shape
```

```python
#No of rows to train model on
train_data_len=math.ceil(len(dataset)*.7)
train_data_len
```

```python
 #Scaling data using Min Max Scaler since LSTM are sensitive to unscaled data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
scaled_data=scaler.fit_transform(dataset)
scaled_data
```

```python
#Data Preprocessing : Various ways to do train test split like Cross Validation , random seed but these only work well when we are doing regression or classification 
#But in case of time series the data should be ordered wrt date since here days are dependendent on previous days
```

```python
#Training data set
train_data=scaled_data[0:train_data_len,:]
```

```python
#Splitting data into train and test 
X_train=[]
y_train=[]

for i in range (60,len(train_data)):
  X_train.append(train_data[i-60:i,0]) #appending past 60 values to X_train ... it will have value from position 0 to position 59
  y_train.append(train_data[i,0]) #it will contain the 61th value(position 60)
  if i<=60:
    print(X_train)
    print(y_train)
    print()

#X_train contains past 60 values
#y_train contains 61th value that our model needs to predict
```

```python
#Creating stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(X_train.shape[1],1))) #First hidden layer is 50 and input shape should be X_train[1] ie (100,1)
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
#model.add(Dense(25))
model.add(Dense(units=1,activation='linear'))
```

```python
#Compiling the model
model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
 #score= model.evaluate(X_test, y_test)
```

```python
#Training model with batc size 
model.fit(X_train,y_train,epochs=100,batch_size=64,validation_split=0.2)
```

```python
#Creating test dataset

test_data=scaled_data[train_data_len-60:,:]
#Creating dataset x_test,y_test
X_test=[]
y_test=scaled_data[train_data_len:,:]
for i in range(60,len(test_data)):
  X_test.append(test_data[i-60:i,0])
```
